# Read

```python
import pandas as pd

train=pd.read_csv('./data/preprocess/train_0926.csv')
test=pd.read_csv('./data/preprocess/test_0926.csv')
```

```python
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score
import numpy as np

# RMSLE 정의
def rmsle(origin, pred):
    return np.sqrt(mean_squared_log_error(origin, pred))
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
train_features=train.drop(['SalePrice'],axis=1)
pca=PCA(n_components=30)

pca.fit(train_features)

train_pca=pca.transform(train_features)
train_pca=pd.DataFrame(train_pca)

test_pca=pca.transform(test)
test_pca=pd.DataFrame(test_pca)
```

```python
train_pca['SalePrice']=train['SalePrice']
```

```python
train_pca.columns=['PCA_1','PCA_2','PCA_3','PCA_4','PCA_5','PCA_6','PCA_7','PCA_8','PCA_9','PCA_10','PCA_11','PCA_12','PCA_13','PCA_14','PCA_15','PCA_16','PCA_17','PCA_18','PCA_19','PCA_20','PCA_21','PCA_22','PCA_23','PCA_24','PCA_25','PCA_26','PCA_27','PCA_28','PCA_29','PCA_30','SalePrice']
test_pca.columns=['PCA_1','PCA_2','PCA_3','PCA_4','PCA_5','PCA_6','PCA_7','PCA_8','PCA_9','PCA_10','PCA_11','PCA_12','PCA_13','PCA_14','PCA_15','PCA_16','PCA_17','PCA_18','PCA_19','PCA_20','PCA_21','PCA_22','PCA_23','PCA_24','PCA_25','PCA_26','PCA_27','PCA_28','PCA_29','PCA_30']
```

# OLS

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(train_pca,
                                       test_size = 0.3,
                                       random_state = 2045)

train_set.shape, test_set.shape
```

```python
test_set.head()
```

```python
import statsmodels.formula.api as smf

Model_1 = smf.ols(formula = 'SalePrice ~ PCA_1+PCA_2+PCA_3+PCA_4+PCA_5+PCA_6+PCA_7+PCA_8+PCA_9+PCA_10+PCA_11+PCA_12+PCA_13+PCA_14+PCA_15+PCA_16+PCA_17+PCA_18+PCA_19+PCA_20+PCA_21+PCA_22+PCA_23+PCA_24+PCA_25+PCA_26+PCA_27+PCA_28+PCA_29+PCA_30',
                  data = train_set).fit()
y_hat_1 = Model_1.predict(test_set[['PCA_1','PCA_2','PCA_3','PCA_4','PCA_5','PCA_6','PCA_7','PCA_8','PCA_9','PCA_10','PCA_11','PCA_12','PCA_13','PCA_14','PCA_15','PCA_16','PCA_17','PCA_18','PCA_19','PCA_20','PCA_21','PCA_22','PCA_23','PCA_24','PCA_25','PCA_26','PCA_27','PCA_28','PCA_29','PCA_30']])
rmsle(np.expm1(test_set['SalePrice']),np.expm1(y_hat_1))
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(Model_1, test_set,'SalePrice',scoring='neg_mean_squared_root_error', cv=5)
```

# Submission

```python
Model = smf.ols(formula = 'SalePrice ~ PCA_1+PCA_2+PCA_3+PCA_4+PCA_5+PCA_6+PCA_7+PCA_8+PCA_9+PCA_10+PCA_11+PCA_12+PCA_13+PCA_14+PCA_15+PCA_16+PCA_17+PCA_18+PCA_19+PCA_20+PCA_21+PCA_22+PCA_23+PCA_24+PCA_25+PCA_26+PCA_27+PCA_28+PCA_29+PCA_30',
                  data = train_pca).fit()
y_hat = Model.predict(test_pca[['PCA_1','PCA_2','PCA_3','PCA_4','PCA_5','PCA_6','PCA_7','PCA_8','PCA_9','PCA_10','PCA_11','PCA_12','PCA_13','PCA_14','PCA_15','PCA_16','PCA_17','PCA_18','PCA_19','PCA_20','PCA_21','PCA_22','PCA_23','PCA_24','PCA_25','PCA_26','PCA_27','PCA_28','PCA_29','PCA_30']])
pred = np.expm1(y_hat)
```

```python
sub=pd.read_csv('data/sample_submission.csv')
sub['SalePrice']=pred
sub.to_csv('./submission/OLS_with_PCA_2.csv',index=False)
```
